3.	What happens if stopwords are not removed?

    ▫️Stopwords are common words like is, the, am, are, in.
    ▫️If not removed adds noise to data, increases computation time, reduces model performance, important words get diluted.
    ▫️For example, consider "This is the best movie".
        Without removal - many unimportant words.
        With removal - "best movie" (more meaningful).

3.	Provide two real-world scenarios where removing stopwords can be harmful.
  
    ▫️Scenario 1: Sentiment Analysis
          Sentence: "I am not happy".
          If stop words are removed, "happy".
          Meaning of the sentence completely changes (negative -> positive)
  
    ▫️Scenario 2: Chatbots
          Question: "What is the time?"
          After removing stopwords, "time".
          Loses structure, system may not understand it's a question.
  
4.	Explain the difference between stemming and lemmatization.

    ▫️Stemming is a technique that removes prefixes/suffixes to get the root form of word, often producing non-meaningful word. (e.g., studies -> studi)
    ▫️Whereas Lemmatization converts a word to its base form using vocabulary and grammar rules, producing meaningful words. (e.g, studies -> study)
    ▫️Stemming is faster but less accurate, while Lemmatization is slower but more accurate.

```python
#-----Task 2: Build an Advanced Preprocessing Function

def preprocess_text(text):
    
    # Error Handling
    if not text or not isinstance(text, str):
        return []
        
    # Convert text to Lowercase
    text = text.lower()

    #Remove URls and emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    #Remove numbers
    text = re.sub(r'\d+', '', text)

    #Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    #Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    #Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    #Tokenisation
    tokens = text.split()
    
    #Remove very short tokens (<2), keep "no", "not"
    cleaned_tokens = [ word for word in tokens if len(word) > 2 or word in ["no", "not"] ]

    return cleaned_tokens

#-----Task 3: Stress Testing

import numpy as np
import re

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length =  np.mean([len(t) for t in tokens]) if tokens else 0

    all_tokens.extend(tokens)

    print("\n---")
    print(" Original:", sentence)
    print(" Tokens:", tokens)
    print(" Cleaned:",  cleaned_sentence)
#-----Task 4: Token Analytics
    print(" Total Tokens:", total_tokens)
    print(" Unique Tokens:", unique_tokens)
    print(" Average Token Length:", round(avg_length, 2))

#-----Task 5: Frequency Analysis

from collections import Counter

counter = Counter(all_tokens)

print("\n Top 10 Most Frequent Words: ")
print(counter.most_common(10))

print("\n TOp 5 Least Frequent Words: ")
print(counter.most_common()[:-6:-1])

#-----Task 6: Build Full Pipeline

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        clean_sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

#-----Task 7: Error Handling ✔️

result = full_pipeline(test_sentences)
print("\n Final Tokens:",  result["tokens"])
print("\n Clean Sentences:", result["clean_sentences"])
```
